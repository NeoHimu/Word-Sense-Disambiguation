```python
import collections
import pickle
```

```python
with open('../glove/vocab_glove', 'rb') as f:
    vocab = pickle.load(f)

with open('train.pickle', 'rb') as f:
    train = pickle.load(f)
```

```python
def word_list(string):
    word_list = []
    count = 0
    for sent_list in train[string][0]:
        word_list.extend(sent_list)
    return word_list

def lower(list_):
    low_list = []
    for string in list_:
        low_list.append(string.lower())
    return low_list

train_words = lower(word_list('hard')) + lower(word_list('interest')) + lower(word_list('line')) + lower(word_list('serve'))
```

```python
print("No. of distinct Words: ", len(set(train_words)))

len(train_words), len(vocab)
```

```python
counter = collections.Counter(train_words)
counter = dict(counter)
counter['interest']
```

```python
unknown_words = []
for word in set(train_words):
    if word not in vocab:
        unknown_words.append(word)
        
un_counter = collections.Counter(unknown_words)
un_counter = dict(un_counter)

sorted_un_counter = sorted(un_counter.items(), key=lambda x:x[1], reverse=True)
sorted_un_counter
```

```python
with open('words_not_in_vocab.pickle','wb') as f:
    pickle.dump(sorted_un_counter, f)
```
